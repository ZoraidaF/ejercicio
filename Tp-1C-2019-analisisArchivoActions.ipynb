```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



%matplotlib inline
```

```python
#subasta
auctions = pd.read_csv('auctions.csv.gzip', compression='gzip') 
```

```python
#Los datos corresponden al anio 2019 y mes 3
auctions['fecha']=pd.to_datetime(auctions['date'])
auctions['day']=auctions['fecha'].dt.day
auctions['hour']=auctions['fecha'].dt.hour
auctions['minute']=auctions['fecha'].dt.minute
auctions['second']=auctions['fecha'].dt.second
```

```python
archivoSubasta=auctions.drop(auctions.columns[[0,1,2,7]], axis='columns')
```

## Análisis 1: Cantidad de participantes que hay en en cada fuente de intercambio (source_id), según la hora y dia 

```python
subasta0=archivoSubasta[archivoSubasta.source_id==0]
cantidadDeParticipantes=subasta0.groupby(['day','hour']).size()
tabla0 = cantidadDeParticipantes.unstack(fill_value=0)
tabla0
```

```python

fig,ax=plt.subplots()
fig.set_size_inches(11.7,8.27)

mapa=sns.heatmap(tabla0 ,cmap='YlOrRd',ax=ax)
mapa.set_title('Cantidad de participantes que participan en source_id=0')
mapa
```
